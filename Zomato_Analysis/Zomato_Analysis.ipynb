**Heatplot**

```python
plt.figure(figsize=(18,10))
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
plt.show()
```

*This heatplot showa the null value present in the dataset*

```python
#Another DataFrame
df_country=pd.read_excel('Country-Code.xlsx')
df_country
```

```python
#merge two DataFrame
final_df=pd.merge(df,df_country,on='Country Code',how='left')
final_df.head(2)
```

**Plot pie chart with respect to Country**

```python
Country_name=final_df.Country.value_counts().index
Country_name
```

```python
Country_values=final_df.Country.value_counts().values
Country_values
```

```python
plt.pie(Country_values[:3],labels=Country_name[:3],autopct='%1.2f%%')
plt.show()
```

*Zomato maximum records or transactions are from India after that USA and then UK.*

```python
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:"Ratting Count"})
ratings
```

## Observation


```python
plt.figure(figsize=(18,5))
fig_1=sns.barplot(x='Aggregate rating',y='Ratting Count',data=ratings)
for bars in fig_1.containers:
    fig_1.bar_label(bars)
plt.show()
```

```python
plt.figure(figsize=(18,5))
fig_2=sns.barplot(x='Aggregate rating',y='Ratting Count',hue='Rating color',data=ratings,palette=['white','red','orange','green','green'])
for bars in fig_2.containers:
    fig_2.bar_label(bars)
plt.show()
```

**Countries that has given 0 rattings**

```python
data=final_df.groupby(['Aggregate rating','Country']).size().reset_index().head(4)
data
```

*Maximum number of 0 rattings are from Indian Customars*
